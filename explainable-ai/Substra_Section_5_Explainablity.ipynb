Les méthodes SHAP et LIME sont les méthodes les plus connues pour aborder l'explicabilité des modèles de Machine Learning.

# **SHAP**

---
L'une des méthodes les plus populaires aujourd'hui, SHAP (SHapley Additive ExPlanations) est une approche basée sur la théorie des jeux pour expliquer la sortie de tout modèle ML. 

Cet algorithme a été publié pour la première fois en 2017 par Lundberg et Lee  et c'est un excellent moyen de rétro-concevoir la sortie de tout algorithme prédictif.

image.png


**Théorie des jeux et apprentissage automatique**

---

Les valeurs SHAP sont basées sur les valeurs Shapley, un concept issu de la théorie des jeux. 

Imaginons que nous ayons un modèle prédictif, alors :
*   le « jeu » reproduit le résultat du modèle,
*   les « joueurs » sont les caractéristiques incluses dans le modèle.

Ce que fait Shapley, c'est quantifier la contribution que chaque joueur apporte au jeu. Ce que fait SHAP, c'est quantifier la contribution que chaque caractéristique apporte à la prédiction faite par le modèle.

Il est important de souligner que ce que nous avons appelé un « jeu » concerne une seule observation. En effet, SHAP concerne l'interprétabilité locale d'un 
modèle prédictif.

Le calcul des valeurs Shapley est d'une complexité importante. SHAP introduit des approximations pour diminuer cette complexité et permettre d'appliquer le concept à des algorithmes ayant de nombreuses features.

# **LIME**
---


LIME a été l'une des premières techniques à gagner en popularité dans le domaine de l'explicabilité. 

LIME signifie Local interpretable model agnostic explanations. 

Actuellement, Lime aide à expliquer les prédictions pour les classificateurs de données tabulaires, d'images et de texte.

Lime essaie essentiellement de donner une approximation linéaire locale du comportement du modèle en créant des modèles de substitution locaux entraînés pour imiter les prédictions du modèle ML localement. 

Dans Interpretable ML, Christoph Molnar décrit LIME avec le process suivant, qui donne une bonne idée intuitive de LIME 
*   Sélectionnez votre instance pour laquelle vous souhaitez avoir une explication de sa prédiction 
*   Perturbez votre jeu de données et obtenez les prédictions pour ces nouveaux points
*   Pondérer les nouveaux échantillons en fonction de leur proximité avec l'instance d'intérêt
*   Former un modèle pondéré et interprétable sur l'ensemble de données avec les variations
*   Expliquer la prédiction en interprétant le modèle local

https://christophm.github.io/interpretable-ml-book/lime.html

Mais LIME a certaines limitations, et doit être utilisé avec précautions, car la définition du voisinage ou de la proximité pour appliquer LIME n'est pas évidente. L'article suivant décrit bien cela : https://christophm.github.io/interpretable-ml-book/lime.html#disadvantages-11
